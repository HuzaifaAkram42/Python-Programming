### ❓ Question 19 — Numeric Literals
**🟡 Medium | Topic: Literals**

Create **four variables**, each representing the number **255**, using different literal formats:
1. Decimal
2. Binary
3. Octal
4. Hexadecimal

Print all four and verify they are all equal using `==`.

````python
decimal_val = 255        # base 10 — no prefix
binary_val  = 0b11111111  # base 2 — prefix 0b
octal_val   = 0o377      # base 8 — prefix 0o
hex_val     = 0xFF       # base 16 — prefix 0x

print(decimal_val, binary_val, octal_val, hex_val)
# output: 255 255 255 255 — Python always prints in decimal

print(decimal_val == binary_val == octal_val == hex_val)
````
