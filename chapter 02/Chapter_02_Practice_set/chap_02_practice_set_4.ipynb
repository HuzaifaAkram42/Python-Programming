### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
p1x1 = int(input("Enter 1st point x coordinate "))
p1y1 = int(input("Enter 1st point y coordinate "))
p2x2 = int(input("Enter 2nd point x coordinate "))
p2y2 = int(input("Enter 2nd point y coordinate "))

d = ((p2x2 - p1x1) ** 2 + (p2y2 - p1y1) **2)**0.5

print("Euclidean Distance ", round(d,3))
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





head 4 legs 12
dogs 2 chicken 2

2l + 4l = 12

````python
# Write your code here
heads = int(input("Enter Number of Heads "))
legs = int(input("Enter Number of Legs "))


# C + D = heads
# 2 * C + 4 * D = legs
C = (4 * heads - legs)//2
D = heads - C

print("Dogs:",D)
print("Chickens:", C)
````

Cars and Bikes

A parking lot contains only cars and bikes.

Car = 4 wheels
Bike = 2 wheels

Input: Total vehicles and total wheels.

Output: Number of cars and bikes.

Example:

Vehicles = 10
Wheels = 28

Cars = 4

````python
vehicles = int(input("Enter Number of Vehicles "))
wheels = int(input("Enter Number of Wheels "))

Car = (wheels - 2 * vehicles) // 2
Bike = vehicles - Car

print("Car:",Car)
print("Bike:",Bike)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

Hint - Thus, the sum of the squares of first n natural numbers =  n(n+1)(2n+1)/6

````python
n = int(input("Enter n "))
sum = n * (n + 1) * (2 * n + 1)/6
print(f"The Sum of the squares of first {n} natural Numbers ",sum)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

Hint - an = a + (n – 1)d

````python
first_term = int(input("Enter first term of the Arithmetic series "))
second_term = int(input("Enter Second term of the Arithmetic series "))
n = int(input("Enter the Nth term "))
d = second_term - first_term
an = first_term + (n - 1) * d

print(f"The {n}th Term of the series",an)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
n1 = int(input("Enter the first number numerator "))
d1 = int(input("Enter the first number denominator "))
n2 = int(input("Enter the second number numerator "))
d2 = int(input("Enter the Second number denominator "))

print(f"""

        {n1}/{d1} + {n2}/{d2}
""")

rn = d2*n1 + d1*n2
rd = d1 * d2

print(f"The result: {rn}/{rd}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
h_t = int(input("Enter height of the milk tank "))
l_t = int(input("Enter lenght of the milk tank "))
b_t = int(input("Enter breadth of the milk tank "))

h_g = int(input("Enter height of the glass "))
r_g = int(input("Enter radius of the glass "))

v_t = h_t * l_t * b_t
v_g = 3.142 * (r_g ** 2) * h_g

print("Total Number of Glasses",int(v_t//v_g))
````
